```python
import pandas as pd
```

```python
movies = pd.read_csv("movie_data.csv")
movies.head()
```

```python
movies = movies.sort_values ascending = False)
```

```python
movies = movies.join(movies['Country'].str.split(',', expand=True).add_prefix('Country'))
```

```python
movies
```

```python
movies.drop(columns='country1')
```

```python
movies = movies.join(movies['Language'].str.split(',', expand=True).add_prefix('Language'))
movies.head()
```

```python
movies = movies.join(movies['Genres'].str.split(',', expand=True).add_prefix('Genres'))
movies.head()
```

```python
movies.to_csv("new_movie.csv")
```
